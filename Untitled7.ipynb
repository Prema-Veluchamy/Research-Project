```python
df = pd.read_csv('/content/Restaurant_Reviews.csv')
```

```python
# Extract text and labels from the dataset
texts = df["Review"].values
labels = df["Liked"].values
```

```python
# Split the data into train and test sets
train_texts, test_texts, train_labels, test_labels = train_test_split(texts, labels, test_size=0.2, random_state=42)
```

```python
# Define batch size and calculate the number of batches
batch_size = 32
num_batches = len(train_texts) // batch_size + 1
```

```python
# Initialize an empty array to store the ELMo embeddings
train_embeddings = []
```

```python
# Process the train data in batches
for i in range(0, len(train_texts), batch_size):
    batch_texts = train_texts[i:i+batch_size]

    # Obtain ELMo embeddings for the batch
    batch_embeddings = elmo(batch_texts)["elmo"]

    # Append the batch embeddings to the overall train embeddings
    train_embeddings.append(batch_embeddings)

# Concatenate the train embeddings along the batch axis
train_embeddings = np.concatenate(train_embeddings, axis=0)

# Flatten the train labels array
train_labels = train_labels.flatten()
```
