```python
import torch
from allennlp.modules.elmo import Elmo, batch_to_ids
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# Specify the ELMo options and weights files
options_file = "https://allennlp.s3.amazonaws.com/models/elmo/2x4096_512_2048cnn_2xhighway/elmo_2x4096_512_2048cnn_2xhighway_options.json"
weight_file = "https://allennlp.s3.amazonaws.com/models/elmo/2x4096_512_2048cnn_2xhighway/elmo_2x4096_512_2048cnn_2xhighway_weights.hdf5"

# Create an instance of the Elmo class
elmo = Elmo(options_file, weight_file, num_output_representations=1)

# Read the dataset using pandas (replace with your own dataset file path)
dataset_path = "your_dataset.csv"
df = pd.read_csv(dataset_path)

# Extract text and labels from the dataset
texts = df["Review"].values
labels = df["Liked"].values

# Split the data into training and test sets
train_texts, test_texts, train_labels, test_labels = train_test_split(texts, labels, test_size=0.2, random_state=42)

# Tokenize the sentences and convert them to character IDs
train_character_ids = batch_to_ids(train_texts)
test_character_ids = batch_to_ids(test_texts)

# Define batch size and calculate the number of batches
batch_size = 32
train_batches = len(train_character_ids) // batch_size + 1
test_batches = len(test_character_ids) // batch_size + 1

# Embed the training data
train_embeddings = []
for i in range(train_batches):
    start_idx = i * batch_size
    end_idx = (i + 1) * batch_size
    character_ids_batch = train_character_ids[start_idx:end_idx].to(device="cuda")
    with torch.no_grad():
        embeddings = elmo(character_ids_batch)["elmo_representations"][0].detach().cpu().numpy()
    train_embeddings.append(embeddings)

X_train_features = np.concatenate(train_embeddings, axis=0)

# Embed the test data
test_embeddings = []
for i in range(test_batches):
    start_idx = i * batch_size
    end_idx = (i + 1) * batch_size
    character_ids_batch = test_character_ids[start_idx:end_idx].to(device="cuda")
    with torch.no_grad():
        embeddings = elmo(character_ids_batch)["elmo_representations"][0].detach().cpu().numpy()
    test_embeddings.append(embeddings)

X_test_features = np.concatenate(test_embeddings, axis=0)

# Train an SVM classifier
svm = SVC()
svm.fit(X_train_features, train_labels)

# Predict labels for test data
predictions = svm.predict(X_test_features)

# Calculate precision, recall, and F-score
precision = precision_score(test_labels, predictions)
recall = recall_score(test_labels, predictions)
f1 = f1_score(test_labels, predictions)

print("Precision:", precision)
print("Recall:", recall)
print("F-score:", f1)
```
