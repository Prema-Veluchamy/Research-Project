```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import OneHotEncoder
from scipy.sparse import hstack

# Sample data
texts = ['This is a sample sentence.',
         'Another sentence for demonstration.',
         'Yet another example sentence.']

pos_tags = ['NOUN', 'VERB', 'ADJ']

# BoW representation of text
vectorizer = CountVectorizer()
bow_features = vectorizer.fit_transform(texts)

# One-hot encoding of POS tags
encoder = OneHotEncoder(sparse_output=False)
pos_encoded = encoder.fit_transform([[pos] for pos in pos_tags])
print(pos_encoded)

# Concatenate BoW features with POS tag encoding
concatenated_features = hstack((bow_features, pos_encoded))

# Print the concatenated features
print(concatenated_features.toarray())
```
